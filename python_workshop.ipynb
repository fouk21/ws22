## Η έννοια της μεταβλητής
Η μεταβλητή αντιπροσωπεύει μία θέση μνήμης (ή πολλές) στον υπολογιστή.
1. Μία μεταβλητή πρέπει να έχει ένα όνομα
2. Το περιεχόμενό της ονομάζεται τιμή της μεταβλητής

Στην python ορίζοντας μία μεταβλητή μπορούμε να της αναθέσουμε όποια τιμή θέλουμε, ανεξαρτήτως τύπου
χρησιμοποιώντας το σύμβολο της ανάθεσης ‘**=**’

````python
# Examples of assigning and printing a variable

# int (integer number)
int_ex = 7
print("int_ex =", int_ex)

# real (real number)
real_ex = 3.14
print("real_ex =", real_ex)

# string (consecutive characters)
string_ex = "This is a string"
print("string_ex =", string_ex)

# boolean (binary values: True or False)
boolean_ex = True
print("boolean_ex =", boolean_ex)
````

````python
# Warning 
# The above variables are called 'global variables'
# and are visible throughout the notebook, check this

print("In different cell from definition\n")

print("int_ex =", int_ex)
print("real_ex =", real_ex)
print("string_ex =", string_ex)
print("boolean_ex =", boolean_ex)

print("\nThose global variables are still here, as thought so")
````

## Τα σύμβολα πράξεων
Στην python μεταξύ των αριθμών ισχύουν οι γνωστές πράξεις των μαθηματικών και ακολουθείται η γνωστή σειρά
προτεραιότητας (παρενθέσεις, πολλαπλασιασμός - διαιρέσεις ,προσθέσεις - αφαιρέσεις):
- πρόσθεση: `+`
- αφαίρεση: `-`
- πολλαπλασιασμός: `*`
- διαίρεση: `/`, πχ 6 / 4 -> 1,5
- ακέραια διαίρεση: `//`, πχ 6 // 4 -> 1
- υπόλοιπο: `%`, πχ 6 % 4 -> 2
- ισότητα: `==`, πχ 2 == 2 -> True
- ανισότητα: `!=`, πχ 2 != 2 -> False

````python
# Examples of operations

# arithmetic operations
print("7 + 5 =", 7 + 5)
print("7 - 5 =", 7 - 5)
print("7 * 5 =", 7 * 5)
print("7 / 5 =", 7 / 5)
print("7 // 5 =", 7 // 5)
print("7 % 5 =", 7 % 5)
print("7 == 5 =", 7 == 5)
print("7 != 5 =", 7 != 5)

# string operations
print("'hello' + ' world' =", 'hello' + ' world')
print("ha * 4 =", 'ha' * 4)
````

## Δομές επανάληψης
Ας υποθέσουμε ότι θέλουμε να εκτελέσουμε ένα σημείο του κωδικά παραπάνω από μία φορά. Τι θα κάνουμε;

## Η έννοια της συνάρτησης (function)

**Στον προγραμματισμό ως συνάρτηση ορίζουμε ένα σύνολο εντολών που έχουν ομαδοποιηθεί μαζί**. 

Όπως και στα μαθηματικά η συνάρτηση μπορεί να δέχεται ως όρισμα μια ή περισσότερες μεταβλητές και ίσως να επιστρέφει κάποια τιμή.

Για τον ορισμό μιας συνάρτησης χρησιμοποιείται η εξής σύνταξη:
```
def function_name(parameter1, parameter2, ...):
    body_of_function
    return something  # return is optional
```

Η συνάρτηση αφού οριστεί μπορεί να κληθεί έτσι:

`function_name(p1, p2, ...)`

και αν επιστρέφει κάποια τιμή, η τιμή επιστροφής μπορεί να αποθηκευτεί στη μεταβλητή result έτσι:

` result = function_name(p1, p2, ...)`

Ας προσπαθήσουμε να υλοποιήσουμε το παράδειγμα με τις κατηγορίες σε μία συνάρτηση για να την καλέσουμε για διαφορετικούς αριθμούς.

````python
# cell to define function

def categorize_number(check_num, print_or_return):
    category = "undefined"

    if check_num < 0:
        category = 'neg'
    elif check_num == 0:
        category = 'zero'
    elif check_num <= 100:
        category = 'pos_max_100'
    else:
        category = 'above_100'

    # output the category if print_res is True
    if print_or_return:
        print("Number", check_num, "has category:", category)
    else:
        return category
````

````python
# cell to show alternative calls of function

# option 1
categorize_number(check_num=-100, print_or_return=True)
categorize_number(100, print_or_return=True)
categorize_number(200, True)

# option 3
cat_ex = categorize_number(300, False) 
print("Number 300 has category:", cat_ex)

# option 4
num_ex = 7522
print("Number", num_ex, "has category:", categorize_number(num_ex, False))
````

## Είσοδος / Έξοδος (Ι / Ο)

1. **Για να πάρουμε τιμές από το πληκτρολόγιο** χρησιμοποιούμε την συνάρτηση **input**, η οποία μας επιστρέφει strings (συμβολοσειρές)
2. **Για να εμφανίσουμε κάτι στην οθόνη** του υπολογιστή χρησιμοποιούμε την συνάρτηση **print**, όπως έχουμε ήδη δει

````python
# Example of input, output

def print_age_based_message(age):
    age_num = int(age)
    if age_num >= 18:
        print("You are an adult now")
    else:
        print("Not an adult yet...")

def simple_adult_detection_1():
    age = input("input your age:")
    print("age given (string):", age)
    print_age_based_message(age)

def simple_adult_detection_2():
    print_age_based_message(input("input your age:"))
    

print("Call simple_adult_detection_1:")
simple_adult_detection_1()

print("\nCall simple_adult_detection_2:")
simple_adult_detection_2()
````
